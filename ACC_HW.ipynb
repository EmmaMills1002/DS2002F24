```python
import pandas as pd
```

```python
acc = pd.read_csv('/content/acc_players-2324F.csv', header=1)
acc.head()
```

**Basic Analysis**

```python
acc['PTS'].sum()
```

```python
most_minutes = acc[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1)
print(most_minutes)
```

```python
top_rebounds = acc[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
print(top_rebounds)
```

**Player Filtering**

```python
most_mp = acc[acc['MP']>500]
```

```python
most_mp.info()
```

```python
top_assists = acc[['Player', 'AST']].sort_values(by='AST', ascending=False).head(1)
print(top_assists)
```

```python
top_three_assists = acc[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
print(top_three_assists)
```

```python
top_shotblockers = acc[['Player', 'STL']].sort_values(by='STL', ascending=False).head(3)
print(top_shotblockers)
```

**School-Based Analysis**

```python
school_points = acc.groupby('School')[['PTS']].sum().sort_values('PTS', ascending=False)
school_points
```

```python
acc.groupby('School')[['AST']].sum().sort_values('AST', ascending=False)
```
